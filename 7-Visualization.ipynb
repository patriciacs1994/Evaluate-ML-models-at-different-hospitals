# Visualize the outcome of the evaluations

## 1. Import the required libraries

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

## 2. Visualize difference on AUROC between live and retrospective evaluation (%)

### 2.1. Prepare the data and corresponding labels

```python
hospitals = ["N", "M", "H", "AVG"]
metrics = ["Delirium-adm", "Delirium-dis", "Sepsis-adm",
           "Sepsis-dis", "AKI-adm", "AKI-dis"]
```

```python
metrics_diff = np.array([[-4.36, 0.31, -0.82, -1.05, 3.15, 3.41],
                    [-1.07, -2.14, 0.36, 0.12, -4, -3.91],
                    [-5.3, -3.63, 4.23, 1.33, -0.07, -0.65],
                        [-3.58,-1.82,1.26,0.13,-0.31,-0.38]])
```

### 2.2. Visualize and save the visualization

```python
fig, ax = plt.subplots(figsize=(16,24))
im = ax.imshow(metrics_diff,cmap="RdYlGn")

# Setting the labels
ax.set_xticks(np.arange(len(metrics)))
ax.set_yticks(np.arange(len(hospitals)))
# labeling respective list entries
ax.set_xticklabels(metrics, fontsize = 25.0)
ax.set_yticklabels(hospitals, fontsize = 25.0)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Creating text annotations by using for loop
for i in range(len(hospitals)):
    for j in range(len(metrics)):
        text = ax.text(j, i, metrics_diff[i, j],
                       ha="center", va="center", color="b", fontsize = 25.0)

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="2%", pad=0.25)

cbar = plt.colorbar(im, cax=cax)
cbar.set_label('Difference on AUROC (%)',labelpad=10,fontsize=20)

cbar.ax.tick_params(labelsize=20) 

ax.set_title("Difference on AUROC between live and retrospective evaluation (%)", fontsize = 25.0)
fig.tight_layout()
plt.savefig('model_performance_comparison.png')
plt.show()
```

## 2. Visualize the performance degradation in cross-hospital evaluation

### 2.1. Performance degradation when a model is deployed to a different hospital other than the hospital it is trained

```python
#performance degradation when a model is deployed to a different hospital other than the hospital it is trained
hospitals = ["Model-H", "Model-M", "Model-N"]
metrics = ["H", "M", "N","H", "M", "N","H", "M", "N"]

metrics_diff = np.array([
[0,-3.2,-9.2,	0,-11.9,-5.6,	0,-20.5,-12.2],
[-9.5,0,-5.1,	-8.3,0,-11.4,	-19.1,0,-8.4],
[-4.1,-0.8,0,	-0.1,-4.5,0,	-1.3,-7.4,0]])
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable


fig, ax = plt.subplots(figsize=(24,36))
im = ax.imshow(metrics_diff,cmap="gist_heat")

# Setting the labels
ax.set_xticks(np.arange(len(metrics)))
ax.set_yticks(np.arange(len(hospitals)))
# labeling respective list entries
ax.set_xticklabels(metrics, fontsize = 25.0)
ax.set_yticklabels(hospitals, fontsize = 25.0)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), ha="right",
         rotation_mode="anchor")


divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="2%", pad=0.25)

cbar = plt.colorbar(im, cax=cax,label='Performance degradation on AUROC (%)')
cbar.set_label('Performance degradation on AUROC (%)',labelpad=10,fontsize=20)
cbar.ax.tick_params(labelsize=20) 


# Creating text annotations by using for loop
for i in range(len(hospitals)):
    for j in range(len(metrics)):
        text = ax.text(j, i, metrics_diff[i, j],
                       ha="center", va="center", color="g", fontsize = 25.0)

fig.tight_layout()
plt.savefig('model_performance_comparison.png')
plt.show()
```
